```python
import pandas as pd; import numpy as np; import matplotlib.pyplot as plt; import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_absolute_percentage_error
```

```python
df = pd.read_excel('/content/sample_data/Graded-assessment-College.xlsx')
df.head()
```

```python
# drop 0 column
df.drop(columns=['Unnamed: 0'], inplace=True)
df.head()
```

```python
df.describe().round(2)
```

```python
if 'Private' in df.columns:
    df = pd.get_dummies(df, columns=['Private'], drop_first=True)
    print("One-hot encoding applied to 'Private' column.")
else:
    print("'Private' column not found, assuming it has already been encoded.")
df.head()
```

```python
x = df.drop(columns=['Grad.Rate'])
y = df['Grad.Rate']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

#Linear Regression Using scikit-learn

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_train_pred = lr.predict(x_train)
y_test_pred = lr.predict(x_test)
```

```python
print(f"Intercept term: {lr.intercept_:.2f}")
```

```python
coefficients = pd.DataFrame({
    'feature': x_train.columns,
    'coefficient': lr.coef_
})

# Identify categorical features - in this case, only 'Private_Yes' after one-hot encoding
categorical_features = ['Private_Yes']

categorical_coefficients = coefficients[coefficients['feature'].isin(categorical_features)]
categorical_coefficients

if not categorical_coefficients.empty:
    highest_coeff_categorical = categorical_coefficients.loc[categorical_coefficients['coefficient'].abs().idxmax()]
    print(f"The categorical predictor with the highest coefficient is '{highest_coeff_categorical['feature']}' with a coefficient of {highest_coeff_categorical['coefficient']:.2f}.")
else:
    print("No categorical features found with coefficients.")
```

```python
categorical_coefficients
```

```python
# Exclude categorical features
numerical_coefficients = coefficients[~coefficients['feature'].isin(categorical_features)]

if not numerical_coefficients.empty:
    highest_coeff_numerical = numerical_coefficients.loc[numerical_coefficients['coefficient'].abs().idxmax()]
    print(f"Excluding categorical predictors, the predictor with the highest coefficient is '{highest_coeff_numerical['feature']}' with a coefficient of {highest_coeff_numerical['coefficient']:.2f}.")
else:
    print("No numerical features found with coefficients.")
```

```python
pd.DataFrame({'Actual Train': y_train, 'Predicted Train': np.round(y_train_pred, 2)}).sample(5, random_state=0) # Training data
```

```python
pd.DataFrame({'Actual Test': y_test, 'Predicted Test': np.round(y_test_pred, 2)}).sample(5, random_state=0) #Testing data
```

```python
def regression_metrics(y_train, y_train_pred, y_test, y_test_pred):
  def compute_metrics(y, y_pred):
    mse = mean_squared_error(y, y_pred)
    rsme = np.sqrt(mse)
    mae = mean_absolute_error(y, y_pred)
    mape = mean_absolute_percentage_error(y, y_pred) * 100
    r2 = r2_score(y, y_pred)
    var = y.var()
    std = y.std()
    mean_val = y.mean()

    return mse, var, rsme, std, mae, mean_val, mape, r2


  metrics_train = compute_metrics(y_train, y_train_pred)
  metrics_test = compute_metrics(y_test, y_test_pred)

  return pd.DataFrame({'Dataset': ['Training', 'Testing'], 'MSE': [metrics_train[0], metrics_test[0]],
                       'Variance': [metrics_train[1], metrics_test[1]],
                       'RMSE': [metrics_train[2], metrics_test[2]],
                       'St-Dev': [metrics_train[3], metrics_test[3]],
                       'MAE': [metrics_train[4], metrics_test[4]],
                       'Mean': [metrics_train[5], metrics_test[5]],
                       'MAPE': [metrics_train[6], metrics_test[6]],
                       'R2': [metrics_train[7], metrics_test[7]]}).set_index('Dataset')
```

```python
df_metrics = regression_metrics(y_train, y_train_pred, y_test, y_test_pred).round(2)
df_metrics
```
