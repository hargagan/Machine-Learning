```python
import numpy as np; import pandas as pd; import matplotlib.pyplot as plt; import seaborn as sns
import numpy as np; import pandas as pd; import matplotlib.pyplot as plt; import seaborn as sns
from sklearn.datasets import load_diabetes
from sklearn.preprocessing import MinMaxScaler  # Scaling
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression, Lasso, Ridge
from sklearn.metrics import mean_squared_error
import warnings; warnings.filterwarnings('ignore')
```

```python
# Load the breast cancer data
diabetes_ds =  load_diabetes()
X = pd.DataFrame(diabetes_ds.data, columns=diabetes_ds.feature_names)
print(X.head())
y = pd.Series(diabetes_ds.target)
print(f'Input shape: {X.shape}')
print(f'Target shape: {y.shape}')
print(f'Feature names: {diabetes_ds.feature_names}')
print(f'Target names: {diabetes_ds.keys()}')
```

```python
# Train-Val-Test split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size = 0.2, random_state = 9001)
features_considered = X_train.columns
print('Dimensions of X_train:', X_train.shape); print('Dimensions of y_train:', y_train.shape)
print('Dimensions of X_test:', X_test.shape); print('Dimensions of y_test:', y_test.shape)
```

```python
# Fit a scaler to the training data and transform all the datasets
scaler = MinMaxScaler()
numerical_columns = X_train.select_dtypes(include = ['int64', 'float64']).columns
X_train[numerical_columns] = scaler.fit_transform(X_train[numerical_columns])
X_test[numerical_columns] = scaler.transform(X_test[numerical_columns])
```

# Unregularized Linear Regression Model


```python
lr_model_1 = LinearRegression()
lr_model_1 = lr_model_1.fit(X_train, y_train)
# Look at the predictive performance of this model
y_pred_lr_list = lr_model_1.predict(X_train[features_considered])
mse = mean_squared_error(y_true = y_train, y_pred = y_pred_lr_list)
lr_mse = np.round(mse, 2)
print('LinearRegression - Training MSE = ', lr_mse)
```

# Lasso Regularization

```python
lr_model_2 = Lasso(alpha=1, random_state=9001)
lr_model_2 = lr_model_2.fit(X_train[features_considered], y_train)

y_pred_lr_a_list = lr_model_2.predict(X_train[features_considered])
mse = mean_squared_error(y_true = y_train, y_pred = y_pred_lr_a_list)
lr_a_mse = np.round(mse, 2)
print('Lasso Regression - Training MSE = ', lr_a_mse)
```

```python
lr_model_3 = Ridge(alpha=1, random_state=9001)
lr_model_3 = lr_model_3.fit(X_train[features_considered], y_train)

y_pred_lr_r_list = lr_model_3.predict(X_train[features_considered])
mse = mean_squared_error(y_true = y_train, y_pred = y_pred_lr_r_list)
lr_r_mse = np.round(mse, 2)
print('Ridge Regression - Training MSE = ', lr_r_mse)
```

```python
pd.DataFrame({"Feature": features_considered, "Coefficients - LR": lr_model_1.coef_, "Coefficients - Lasso": lr_model_2.coef_, "Coefficient - Ridge": lr_model_3.coef_}).round(3)
```

```python
pd.DataFrame({"Feature": ["MSE"], "Linear-Regression": lr_mse, "Lasso-LR": lr_a_mse, "Ridge-LR": lr_r_mse}).round(2)
```
